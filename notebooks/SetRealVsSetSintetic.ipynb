```python
real_means = train_data.mean(numeric_only=True)
synthetic_means = synthetic_data.mean(numeric_only=True)

comparison_df = pd.DataFrame({
    "Real": real_means,
    "Synthetic": synthetic_means
})
```

```python
comparison_df.plot(kind='bar', figsize=(12, 8))
plt.title("Comparație valori medii - Set real vs Set sintetic CTGAN")
plt.xlabel("Caracteristici")
plt.ylabel("Valoare medie")
plt.xticks(rotation=90)
plt.grid(axis='y')
plt.tight_layout()
plt.legend(title="Set de date")
plt.show()
```

```python
real_std = train_data.std(numeric_only=True)
synthetic_std = synthetic_data.std(numeric_only=True)

std_comparison_df = pd.DataFrame({
    "Real Std": real_std,
    "Synthetic Std": synthetic_std
})
```

```python
std_comparison_df.plot(kind='bar', figsize=(12, 8))
plt.title("Comparație deviații standard - Set real vs Set sintetic CTGAN")
plt.xlabel("Caracteristici")
plt.ylabel("Deviație standard")
plt.xticks(rotation=90)
plt.grid(axis='y')
plt.tight_layout()
plt.legend(title="Set de date")
plt.show()
```
